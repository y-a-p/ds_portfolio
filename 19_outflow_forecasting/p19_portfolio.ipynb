# Телеком. Отток клиентов

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Ваша задача — обучить на этих данных модель для прогноза оттока клиентов.

## Описание услуг
Оператор предоставляет два основных типа услуг:
1. Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
2. Интернет. Подключение бывает двух типов:
 - через телефонную линию DSL (англ. digital subscriber line — «цифровая абонентская линия»)
 - или оптоволоконный кабель (англ. fiber optic).

Также абонентам доступен ряд услуг:
 - Интернет-безопасность:
  - антивирус (Device Protection)
  - блокировка опасных сайтов (Online Security);
 - Выделенная линия технической поддержки (Tech Support);
 - Облачное хранилище файлов для резервного копирования данных (Online Backup);
 - Стриминговое телевидение (Streaming TV) и каталог фильмов (Streaming Movies).

За услуги клиенты могут платить ежемесячно или раз в 1–2 года. Доступны различные способы расчёта и возможность получить электронный чек.

## Описание данных
Данные хранятся в базе данных PostgreSQL. Она состоит из нескольких таблиц:
 - `contract` — информация о договорах;
 - `personal` — персональные данные клиентов;
 - `internet` — информация об интернет-услугах;
 - `phone` — информация об услугах телефонии.

### Таблица `telecom.contract`
 - `customerID` — ID абонента;
 - `BeginDate` — дата начала действия договора;
 - `EndDate` — дата окончания действия договора;
 - `Type` — тип оплаты: раз в год-два или ежемесячно;
 - `PaperlessBilling` — электронный расчётный лист;
 - `PaymentMethod` — тип платежа;
 - `MonthlyCharges` — расходы за месяц;
 - `TotalCharges` — общие расходы абонента.

### Таблица `telecom.personal`
 - `customerID` — ID пользователя;
 - `gender` — пол;
 - `SeniorCitizen` — является ли абонент пенсионером;
 - `Partner` — есть ли у абонента супруг или супруга;
 - `Dependents` — есть ли у абонента дети.

### Таблица `telecom.internet`
 - `customerID` — ID пользователя;
 - `InternetService` — тип подключения;
 - `OnlineSecurity` — блокировка опасных сайтов;
 - `OnlineBackup` — облачное хранилище файлов для резервного копирования данных;
 - `DeviceProtection` — антивирус;
 - `TechSupport` — выделенная линия технической поддержки;
 - `StreamingTV` — стриминговое телевидение;
 - `StreamingMovies` — каталог фильмов.

### Таблица `telecom.phone`
 - `customerID` — ID пользователя;
 - `MultipleLines` — подключение телефона к нескольким линиям одновременно.

Объеденим таблицы в одну. Customer и personal объеденим с помощью inner, и затем с помощью left добавим internet и phone

```python
df_combined = pd.merge(df_contract, df_personal, how='inner', on=['customerID'])
df_combined.info()
```

```python
df_combined = pd.merge(df_combined, df_internet, how='left', on=['customerID'])
df_combined.info()
```

```python
df_combined = pd.merge(df_combined, df_phone, how='left', on=['customerID'])
df_combined.info()
```

Заполним пропуски.

```python
df_combined[columns_internet] = df_combined[columns_internet].fillna("Not_connected")
```

```python
df_combined['MultipleLines'] = df_combined['MultipleLines'].fillna("Not_connected")
```

Выделим таргет

```python
df_combined['target'] = df_combined['EndDate'].notnull().astype(int)
```

```python
date = pd.to_datetime('01-02-2020', format='%d-%m-%Y')
```

```python
df_combined['duration'] = df_combined.apply(lambda row:
                                            (date if pd.isnull(row['EndDate'])
                                                   else row['EndDate']
                                            ) - row['BeginDate'], axis=1).dt.days
```

## Обучение моделей

```python
X = df_combined.drop('target', axis=1)
y = df_combined['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, stratify=y, random_state = RANDOM_STATE)
```

Выбор метрики. Так как для нас важно разработать модель, которая бы позволяла удерживать как можно больше клиентов, то на мой взгляд более рациональным было бы использовать  метрику recall. Но в процессе обучения моделей я столкнулся с том, что не могу достичь требуемого уровня roc_auc. Поэтому я отказался от recall и состредоточился на roc_auc, как того требовало задание.

```python
class_counts = Counter(y_train)
most_common_class = class_counts.most_common(1)[0][0]
baseline_accuracy = class_counts[most_common_class] / len(y_train)
print("accuracy base line:", baseline_accuracy)
```

```python
preprocessor = ColumnTransformer(
  transformers=[
    ('numeric', StandardScaler(), numeric_columns),
    ('categorical', OrdinalEncoder(), categorical_columns)
  ])
```

### DecisionTree

```python
pipeline_dtc = Pipeline([
  ('preprocessor', preprocessor),
  ('classifier', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
param_grid_dtc = {
  #'classifier__max_depth': [20, 30],
  'classifier__max_depth': [None] + [i for i in range(2, 7)],
  'classifier__min_samples_split': [125, 150],
  'classifier__min_samples_leaf': [20, 30],
  'classifier__criterion': ['gini', 'entropy']
}
```

```python
grid_search_dtc = GridSearchCV(pipeline_dtc, param_grid_dtc, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_dtc.fit(X_train, y_train)

print("Лучшие параметры:", grid_search_dtc.best_params_)
print("Лучшая оценка roc_auc на валидации:", grid_search_dtc.best_score_)
```

### Catboost

```python
pipeline_cb = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', CatBoostClassifier(random_seed=RANDOM_STATE, verbose=False))
])

param_grid_cb = {
  'classifier__n_estimators': [1000, 1500],
  'classifier__learning_rate': [0.01, 0.1],
  'classifier__depth': [4, 6],
  'classifier__border_count': [32, 64],
  'classifier__l2_leaf_reg': [5],
  'classifier__random_strength': [1],
  'classifier__bagging_temperature': [0.5, 1],
}

grid_search_cb = GridSearchCV(pipeline_cb, param_grid_cb, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_cb.fit(X_train, y_train)

print("Лучшие параметры:", grid_search_cb.best_params_)
print("Лучшая оценка roc_auc на обучающем наборе:", grid_search_cb.best_score_)
```

### Net

```python


X_train_transformed = preprocessor.fit_transform(X_train)
input_shape = X_train_transformed.shape[1]

kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

roc_auc_scores = []

for train_index, val_index in kf.split(X_train_transformed):
  X_train_fold, X_val_fold = X_train_transformed[train_index], X_train_transformed[val_index]
  y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

  model = Sequential([
    Dense(128, activation='relu', input_shape=(input_shape,)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(32, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
  ])

  model.compile(optimizer=Adam_Keras(lr=0.001),
                  loss='binary_crossentropy',
                  metrics=[Recall()])

  early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

  model.fit(X_train_fold, y_train_fold, epochs=1000, batch_size=64, validation_data=(X_val_fold, y_val_fold),
              callbacks=[early_stopping, ], verbose=0)

  y_pred_val = model.predict(X_val_fold).ravel()
  roc_auc = roc_auc_score(y_val_fold, y_pred_val)
  roc_auc_scores.append(roc_auc)

print(f"Средний ROC-AUC на кросс-валидации: {np.mean(roc_auc_scores):.4f}")
```

**Вывод**
 - в качестве метрики выбрана roc_auc
 - обучено три модели:  DecisionTree (0.861), Catboost (0.880) и нейронная сеть (0.872)
 - в качестве модели для тестирования выберем Catboost

- False Positive (FP) - 105: это количество клиентов, которых модель неверно классифицировала как "ушедших", хотя на самом деле они остались. Такие ошибки могут привести к ненужным затратам на удержание клиентов, которые и так не собирались уходить.
- False Negative (FN) - 228: это количество клиентов, которых модель неверно классифицировала как "не ушедших", хотя на самом деле они ушли. Это упущенные возможности удержания. Клиенты, которые ушли, не были определены как рискованные, и поэтому не были предприняты меры для их удержания.
- True Positive (TP) - 239: это количество клиентов, которых модель правильно идентифицировала как "ушедших". Это позволяет компании целенаправленно работать с этой группой для попытки изменить их решение об уходе через специальные предложения или лояльные программы.

##### `rating`

Значения представлены в разных форматах: рейтинг может быть записан как 6.2, так и 62%.<br>
Из двузначных значений уберем % и разделим на 10.

```python
df['ratings'].sort_values().unique()
```

```python
def rating_correction(val):
    if type(val) == str and '%' in val:
        return int(val.replace('%', ''))/10
    else:
        return val
```

```python
df['ratings'] = df['ratings'].apply(rating_correction)
```

```python
df['ratings'] = pd.to_numeric(df['ratings'])
```

```python
def split(row):
    try: 
        row['director_main'] = row['director'].split(',')[0]
    except:
        row['director_main'] = row['director']
        
    try: 
        row['genres_main'] = row['genres'].split(',')[0]
    except:
        row['genres_main'] = row['genres']
    
    return row[['director_main','genres_main']]
```
